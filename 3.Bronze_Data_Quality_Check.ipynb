```python
from pyspark.sql.functions import *

def null_check(df, table_name):
    total_rows = df.count()
    
    null_counts = df.select([
        #  for each column, count nulls and calculate percentage
            (count(when(col(c).isNull(), 1)) / total_rows * 100).alias(c) for c in df.columns
        
    ]).collect()[0]
    
    print(f"\n=== {table_name} ===")
    for c, v in zip(df.columns, null_counts):
        print(f"{c}: {v:.2f}%")
```

```python
tables = {
    "orders": "olist_project.bronze.orders",
    "customers": "olist_project.bronze.customers",
    "order_items": "olist_project.bronze.order_items",
    "payments": "olist_project.bronze.payments",
    "reviews": "olist_project.bronze.reviews",
    "products": "olist_project.bronze.products"
}

for name, path in tables.items():
    null_check(spark.table(path), name)
```
